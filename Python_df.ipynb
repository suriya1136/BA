# Intro to SeaBorn visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 5  # Set default figure size
```

```python
#distribution plot
sns.displot(stats['InternetUsers'], bins = 30)  # Distribution plot for 'InternetUsers'
```

```python
#distribution plot
sns.displot(stats['BirthRate'], bins = 30)  # Distribution plot for 'BirthRate'
```

```python
#boxplot
sns.boxplot(data=stats, x='IncomeGroup', y='InternetUsers')  # Boxplot for 'InternetUsers' by 'IncomeGroup'
```

```python
# Scatter plot with regression line
sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', 
           hue='IncomeGroup', # Color by 'IncomeGroup'
           fit_reg=False, # Disable regression line
           scatter_kws={"s":100})  #unwrapped from matplotlib
```
